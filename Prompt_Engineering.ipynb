```python
from openai import OpenAI

client = OpenAI(api_key=api_key1)

user_input = "What is Data Engineering?"

prompt = """
Use the below examples as a template for answers.

Example 1:

Question: What is Machine Learning?

Answer:
- Introduction: Machine Learning is a branch of Artificial Intelligence that enables systems to learn from data.
- History: The concept originated in the 1950s and gained popularity with increased computing power.
- Example: Predicting house prices using historical sales data.

Example 2:

Question: What is Data Science?

Answer:
- Introduction: Data Science is the field of extracting insights from structured and unstructured data.
- History: It evolved from statistics, mathematics, and computer science.
- Example: Customer churn prediction using historical customer behavior.

Example 3:

Question: What is Cloud Computing?

Answer:
- Introduction: Cloud Computing provides on-demand access to computing resources over the internet.
- History: Cloud services became mainstream in the mid-2000s with providers like AWS.
- Example: Hosting a web application on AWS EC2.

Now answer the following question in the same format.

Question: """ + user_input

response = client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[
        {
            "role": "user",
            "content": prompt
        }
    ]
)

print(response.choices[0].message.content)
```
